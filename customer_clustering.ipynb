Dt_Customer (date of customer's enrollment) is not a useful attribute in date form. Therefore, it is converted into the form of "days past since the enrollment".

```python
from datetime import date
from datetime import datetime

today = date.today()
today
```

```python
days_list = []

for i in range(len(data)):
    st = data['Dt_Customer'].values[i]
    str_date = str(today - datetime.strptime(st, '%d-%m-%Y').date())
    c = str_date.find("days")
    str_date = str_date[0:c-1]
    days_list.append(int(str_date))


days_data = pd.DataFrame(days_list)
days_data.T
```

```python
data['Dt_Customer'] = days_data[0]
data_table = pd.DataFrame(data['Dt_Customer'])
data_table.T 
```

```python
#Silhoutte method to find the optimal number of clusters
from sklearn.metrics import silhouette_score

kmeans_per_k = [KMeans(n_clusters=k, random_state=42).fit(pca)
                for k in range(2, 10)]
silhouette_scores = [silhouette_score(pca, model.labels_)
                     for model in kmeans_per_k[1:]]

silhouette_scores
```

```python
plt.figure(figsize=(8, 3))
plt.plot(range(2, 9), silhouette_scores, "bo-")
plt.xlabel("$k$")
plt.ylabel("Silhouette score")
plt.grid()
plt.show()
```

```python
best_k = kmeans_per_k[np.argmax(silhouette_scores)]
best_k
```
